## How Family Income Affects Academics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('graduation_rate.csv')
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['high school gpa'], df['parental income'], color='skyblue')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, y='parental income', x='high school gpa', marker='o', color='green')
plt.title('Parental Income vs High School GPA')
plt.xlabel('High School GPA')
plt.ylabel('Parental Income')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='college gpa', y='parental income', marker='o', color='orange')
plt.title('Parental Income vs College GPA')
plt.xlabel('College GPA')
plt.ylabel('Parental Income')
```

```python
less_than_50000df = df[df['parental income'] < 50000]
```

```python
less_than_50000df
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=less_than_50000df, y='parental income', x='high school gpa', marker='o', color='green')
plt.title('Parental Income vs High School GPA')
plt.xlabel('High School GPA')
plt.ylabel('Parental Income')
```

```python
plt.figure(figsize=(12, 6))
sns.catplot(data=df, x='ACT composite score', y='parental income', kind='box', height=6, aspect=1.5)
plt.title('Parental Income vs ACT Composite Score')
plt.xlabel('ACT Composite Score')
plt.ylabel('Parental Income')
plt.show()
```

```python
df['SAT total score'] = df['SAT total score'] / 1.5
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='SAT total score', y='parental income', alpha=0.4)
plt.title("Parental Income vs SAT Total Score")
plt.xlabel("SAT Total Score")
plt.ylabel("Parental Income")
plt.show()

```

```python
plt.figure(figsize=(10,6))
sns.regplot(
    data=df,
    x='SAT total score',
    y='parental income',
    scatter_kws={'alpha':0.3},   # make points semi-transparent
    line_kws={'color':'red'}     # make regression line stand out
)
plt.title("Parental Income vs SAT Total Score with Regression Line")
plt.xlabel("SAT Total Score")
plt.ylabel("Parental Income")
plt.show()
```

```python
m, b = np.polyfit(df['parental income'], df['SAT total score'], 1)
print(f"Slope: {m}, Intercept: {b}")
```

```python
print(f"Slope per $10000: {round(m * 10000, 2)}")
```
